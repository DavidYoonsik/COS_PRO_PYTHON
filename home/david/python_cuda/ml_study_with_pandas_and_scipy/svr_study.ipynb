# Encode / Normalization

````python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from numpy import concatenate
````

````python
# dataset20.drop('day', axis=1, inplace=True)

# M002	M017	M001	M039	error-count
dataset20.drop('M002', axis=1, inplace=True)
dataset20.drop('M017', axis=1, inplace=True)
dataset20.drop('M001', axis=1, inplace=True)
dataset20.drop('M039', axis=1, inplace=True)
# dataset20.drop('error', axis=1, inplace=True)
````

````python
# Numpy array
value = dataset20.values
````

````python
# Data type
value = value.astype('float64')
````

````python
# 0 ~ 1 scalling
scaler = MinMaxScaler(feature_range=(0, 1))
scaled = scaler.fit_transform(value)
````

````python
scaled
````

# SVR 학습 데이터 분리

````python
# 학습 데이터
xTrain = scaled[:-5000, 1:]
yTrain = scaled[:-5000, 0]

# 테스트 데이터
xTest = scaled[-5000:, 1:]
yTest = scaled[-5000:, 0]
````

# SVR 학습

````python
# 학습
clf = SVR(kernel='linear')
model = clf.fit(xTrain, yTrain)
````

````python
# 예측
preds = model.predict(xTest)
````

# Matplotlib

````python
origin_yTest = yTest.reshape(-1, 1)
origin_yTest = concatenate((origin_yTest, xTest[:, -97:]), axis=1)
origin_yTest = scaler.inverse_transform(origin_yTest)
origin_yTest = origin_yTest[:, 0]

origin_preds = preds.reshape(-1, 1)
origin_preds = concatenate((origin_preds, xTest[:, -97:]), axis=1)
origin_preds = scaler.inverse_transform(origin_preds)
origin_preds = origin_preds[:, 0]

pyplot.figure(figsize=(18,8))
pyplot.plot(origin_yTest)
pyplot.plot(origin_preds)
pyplot.show()
````

# 학습 훈련도 검증

````python
from sklearn.metrics import mean_squared_error
from math import sqrt
````

````python
rmse = sqrt(mean_squared_error(origin_yTest, origin_preds))
print('Test RMSE: %.3f' % rmse)
````

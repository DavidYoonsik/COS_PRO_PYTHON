# Encode / Normalization


```python
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from numpy import concatenate
```

```python
# Numpy array
value = dataset20.values
# value = dataset.values
values_bak = value
```

```python
values_bak
```

```python
day_col = dataset.columns.get_loc('day')
# day_col = 1
encoder = LabelEncoder()
value[:,day_col] = encoder.fit_transform(value[:,day_col])
```

```python
value
```

```python
# Data type
value = value.astype('float32')
```

```python
# 0 ~ 1 scalling
scaler = MinMaxScaler(feature_range=(0, 1))
# Data(bps) transformation 0 to 1
scaled = scaler.fit_transform(value)
```

```python
scaled.shape
```

```python
scaled
```

```python
import numpy as np
# 소수점 1번째 자리까지 반올림
scaled = np.round(scaled, decimals=3)
# scaled2 = scaled.round(4)
```

```python
scaled
```

# SVR 학습 데이터 분리

```python
# 학습 데이터
train_ratio = int(scaled.shape[0]*0.7)

xTrain = scaled[:train_ratio, 1:]
yTrain = scaled[:train_ratio, 0]

# 테스트 데이터
xTest = scaled[train_ratio:, 1:]
yTest = scaled[train_ratio:, 0]
```

# SVR 학습


```python
# 학습
clf = SVR(kernel='linear')
model = clf.fit(xTrain, yTrain)
```

```python
# 예측
preds = model.predict(xTest)
```

# RFR 학습

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor(n_estimators=500, random_state=1)
```

```python
model.fit(xTrain, yTrain)
```

```python
print(model.feature_importances_)
```

```python
names = dataset20.columns.values[0:-1]
ticks = [i for i in range(len(names))]
pyplot.figure(figsize=(18,8))
pyplot.bar(ticks, model.feature_importances_)
pyplot.xticks(ticks, names)
pyplot.show()
```

```python
# 예측
preds = model.predict(xTest)
```

# Matplotlib


```python


origin_yTest = yTest.reshape(-1, 1)
origin_yTest = concatenate((origin_yTest, xTest[:, -1240:]), axis=1)
origin_yTest = scaler.inverse_transform(origin_yTest)
origin_yTest = origin_yTest[:, 0]

origin_preds = preds.reshape(-1, 1)
origin_preds = concatenate((origin_preds, xTest[:, -1240:]), axis=1)
origin_preds = scaler.inverse_transform(origin_preds)
origin_preds = origin_preds[:, 0]

pyplot.figure(figsize=(18,8))
pyplot.plot(origin_yTest)
pyplot.plot(origin_preds)
pyplot.show()
```

```python
from sklearn import neighbors
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
print xTrain.shape
```

```python
print yTrain.shape
```

```python
n_neighbors = 2

for i, weights in enumerate(['uniform', 'distance']):
    knn = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
    y_ = knn.fit(xTrain, yTrain).predict(xTest)

    plt.subplot(2, 1, i + 1)
    plt.scatter(xTest, y_, c='k', label='data')
    plt.plot(xTest, y_, c='g', label='prediction')
    plt.axis('tight')
    plt.legend()
    plt.title("KNeighborsRegressor (k = %i, weights = '%s')" % (n_neighbors, weights))
```

# 학습 훈련도 검증

```python
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
rmse = sqrt(mean_squared_error(origin_yTest, origin_preds))
print('Test RMSE: %.3f' % rmse)
```

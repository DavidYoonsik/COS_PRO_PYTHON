```python
solar_data = pd.read_csv('./data/sample_wetherData.csv')
```

```python
solar_data = solar_data.drop(columns='일시')
```

```python
solar_data.head(20)
```

```python
solar_data['@timestamp'].iloc[0]
```

```python
for i, v in enumerate(solar_data['@timestamp']):
#     print(solar_data['@timestamp'][i])
#     l, r = v.split("  ")
    solar_data['@timestamp'][i] = "2018-"+v
```

```python
solar_data.head()
```

```python
solar_data.index = pd.to_datetime(solar_data['timestamp'])
solar_data.drop('timestamp', axis=1, inplace=True)
```

```python
solar_data.head(20)
```

```python
solar_data = solar_data.resample('15min', how='mean')
```

```python
solar_data.head()
```

```python
import json
from elasticsearch import Elasticsearch
from elasticsearch import helpers
```

```python
data = solar_data.to_dict(orient='records')
```

```python
data[2]['timestamp']
```

```python
es = Elasticsearch(
    ['192.168.2.12:9200']
)
```

```python
request_body = {
    "mappings" : {
        "_doc":{
            "properties":{
                "timestamp":{
                     "type": "date"
                 },
            }
            
         }
    }
}
```

```python
res = es.indices.create(index='solar-energy-for-ml', body=request_body)
```

```python
outer_list = []
for i in data:
    xx = datetime.datetime.strptime(i['timestamp'].strip(), '%m-%d %H:%M:%S').replace(year=2018)
    i['timestamp'] = xx
    action = {"_index": "solar-energy-for-ml", "_type": "_doc", '_source': i}
    outer_list.append(action)
```

```python
outer_list
```

```python
helpers.bulk(es, outer_list)
```
